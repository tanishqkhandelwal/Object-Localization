```python
 
# OpenCV program to perform Edge detection
# import libraries of python OpenCV 

import cv2 

# np is an alias pointing to numpy library 
import numpy as np 

#pd is an alias pointing to pandas library
import pandas as pd

#Reading the test data file
file=pd.read_csv(r'C:\Users\TK\Desktop\contest flipkarrt\day 0\test.csv',header=0)
answer=[]
col=list(file.iloc[:,0])
for image in col:
    flocation=r"C:\Users\TK\Desktop\contest flipkarrt\day 0\images\%s"%( image)
# capture caps from a camera 
    dst = cv2.imread(flocation)  
    cap = cv2.fastNlMeansDenoisingColored(dst,None,10,10,7,21)


# c\test.csvonverting BGR to HSV 
    hsv = cv2.cvtColor(cap, cv2.COLOR_BGR2HSV) 

# define range of red color in HSV 
    lower_red = np.array([30,150,50]) 
    upper_red = np.array([255,255,180]) 

# create a red HSV colour boundary and 
# threshold HSV image 
    mask = cv2.inRange(hsv, lower_red, upper_red) 

# Bitwise-AND mask 
    res = cv2.bitwise_and(cap,cap, mask= mask) 

# finds edges in the input image image and 

    edges = cv2.Canny(cap,100,200)
    
    median_edges = np.median(edges)

    non_empty_columns = np.where(edges.max(axis=0)>median_edges)[0]
    non_empty_rows = np.where(edges.max(axis=1)>median_edges)[0]

    if len(non_empty_columns)==0 or len(non_empty_rows)==0:
        non_empty_rows=[0]
        non_empty_columns=[0]
       
    answer.append([image,min(non_empty_columns), max(non_empty_columns),min(non_empty_rows), max(non_empty_rows)])    
    df = pd.DataFrame(answer, columns = ['image_name', 'x1','x2','y1','y2']) 

df.to_csv('test.csv',index=False)










```
